```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/uybor.xlsx')
df.head()
```

```python
sns.lineplot(x=df['district'], y=df['price'], marker='h', color='red')
plt.xlabel('Tuman')
plt.ylabel("Narx")
plt.grid(alpha=0.5, color='green')
plt.xticks(rotation=90)
plt.title("Tumanlarning narxiga bog'liq grafigi")
plt.show()
```

```python
sns.barplot(x=df['district'], y=df['price'])
plt.title("Uy manzilining narxiga bog'liq grafigi")
plt.xlabel('Manzil')
plt.ylabel('Narx')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='size', y='price', hue='district', style='rooms', s=100)
plt.xticks(rotation=90)
plt.xlabel("O'lcham")
plt.ylabel('Narx')
plt.title('Narx vs Tuman')
plt.show()
```

```python
sns.regplot(data=df, x='size', y='rooms')
plt.show()
```

```python
df_short = df[df['price']<400000]
sns.displot(data=df_short, x='price', hue='district', multiple='stack', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df_short, x='price', kind='kde', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df_short, x='price', kde=True, height=5, aspect=2)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df_short, x='price', hue='district', shade=True)
plt.show()
```

```python
sns.displot(data=df_short, x='price', kind='kde', hue='district', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df_short, x='price', kind='ecdf', height=5, aspect=2)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,10))
plt.suptitle('Grafik 1')
sns.histplot(ax=ax[0], data=df_short, x='price')
sns.scatterplot(ax=ax[1], data=df_short, x=df['price'], y=df['rooms'], hue='district', s=50)
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(12,10))

plt.suptitle('Grafik 2')
sns.histplot(ax=ax[0,0], data=df_short, x='price')
sns.histplot(ax=ax[0,1], data=df_short, x='rooms')
sns.scatterplot(ax=ax[1,0], data=df_short, x='district', y='price')
sns.scatterplot(ax=ax[1,1], data=df_short, x='size', y='price')


ax[0,0].set_title('Price Distribution')
ax[0,1].set_title('Room Distribution')
ax[1,0].set_title('Size vs Price Correlation')
ax[1,1].set_title('District vs Price Correlation')

plt.xticks(rotation=90)
plt.show()
```
