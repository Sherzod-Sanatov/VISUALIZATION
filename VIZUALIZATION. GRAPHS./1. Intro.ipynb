# Common Data Visualization Techniques in Data Science

**Line Chart:**

Purpose: Show trends or changes over time.

When Effective: Best for continuous data, especially time series, to highlight trends, growth, or decline.

**Scatter Plot:**

Purpose: Display relationships or correlations between two continuous variables.

When Effective: Useful for identifying patterns, clusters, or outliers in bivariate data.

**Heatmap:**

Purpose: Represent data in a matrix format using color to indicate values.

When Effective: Best for displaying large datasets or correlations between multiple variables, particularly in identifying patterns or clusters.

**Bar Chart:**

Purpose: Compare discrete categories or groups.

When Effective: Ideal for visualizing categorical data, especially when comparing quantities across different categories.
