```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/refs/heads/main/uzbekistan.csv', index_col=0)
df
```

```python
plt.figure(figsize=(13,6))
sns.barplot(x=df.index, y=df['Aholisi'])
plt.title("O'zbekiston aholisi viloyatlar kesimida")
plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(13,6))
sns.barplot(x=df.index, y=df['Maydoni (kv.km)'])
plt.title("O'zbekiston hududi viloyatlar kesimida")
plt.xticks(rotation=90)

plt.show()
```
