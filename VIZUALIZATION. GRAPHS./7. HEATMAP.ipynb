```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/refs/heads/main/museum_visitors.csv', index_col=0, parse_dates=[0], usecols=[0,1])
df.head()
```

```python
df['month'] = [i.month for i in df.index]
df['year'] = [i.year for i in df.index]
df.head()
```

```python
df2 = df.groupby(['month','year']).max()
df2.head()
```

```python
df2.unstack(level=0)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df2.unstack(level=0), cmap='YlGnBu', linewidths=1, square=True)
plt.show()
```
