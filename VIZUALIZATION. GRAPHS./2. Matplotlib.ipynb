# MATPLOTLIB

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It's widely used for data visualization, especially in fields like data science and machine learning, because it provides fine control over every aspect of a plot. Matplotlib works by allowing users to create a figure (a canvas) and plot various types of graphs (like line, bar, scatter, etc.) on it.

changes we can make

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
x
```

```python
plt.plot(x,y, linewidth=2.5, linestyle='--', color='#8E44AD', marker='o')
plt.title('sin(x) function')
plt.xlabel('PI')
plt.ylabel('sin(x)')
plt.grid(linestyle='--', color='k')

plt.figure(figsize=(30,6))

plt.show()
```

figures on top of each other

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12,6))
# sin(x) graph
plt.plot(x,sin,
         linewidth=2,
         linestyle='--',
         marker='h',
         alpha=0.7,
         color='#8E44AD')
#cos(x) graph
plt.plot(x,cos,
         linewidth=2.0,
         linestyle='--',
         color='c',
         alpha=0.7,
         marker='v')

plt.title("sin(x) va cos(x) grafigi")
plt.xlabel('PI')
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--')
#save an image
plt.savefig('graph1')
plt.show()
```

Grafiklarning turli parametrlarini o'zgartirish yordamida uning tashqi ko'rinishini ham o'zgartirish mumkin:

linewidth - chiziq qalinligi (0.5 dan 3 gacha)

linestyle - chiziq stili

"-" - oddiy chiziq

"--" - yuliq chiziq

"-." - chiziq va nuqta

":" - nuqtali chiziq

None yoki '' - chiziq yo'q

color - chiziq rangi

'b' - ko'k (blue)

'c' - havorang (cyan)

'g' - yashil (green)

'k' - qora (black)

'm' - pushti (magenta)

'r' - qizil (red)

'y' - sariq (yellow)

Bundan tashqari ranglar HEX ('#FF5733') yoki RGBA (255,182,193,0.5) formatlarida ham berilishi mumkin*.

alpha - chiziq shaffofligi (0 shaffof, 0.5 yarim-shaffof, 1 shaffof emas)

marker - marker

'.' - nuqta

'o' - aylana

'v' - pastga qaragan uchburchak

'^' - tepaga qaragan uchburchak

'*' - yulduz

markerfacecolor - markerning rangi
