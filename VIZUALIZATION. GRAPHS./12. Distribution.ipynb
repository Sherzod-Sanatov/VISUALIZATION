```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/refs/heads/main/merc.csv')
df = df[df.year>2010]
df = df[df.fuelType!='Other']
df = df[df.transmission!='Other']
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='price')
plt.show()
```

```python
sns.displot(data=df, x='price', hue='transmission', multiple='stack', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price', col='transmission', height=5, aspect=2)
plt.show()
```

kde - Kerney Density Estimation

```python
sns.displot(data=df, x='price', kind='kde', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price', kde=True, height=5, aspect=2)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='price', hue='fuelType', shade=True)
plt.show()
```

```python
sns.displot(data=df, x='price', kind='kde', hue='fuelType', height=5, aspect=2)
plt.show()
```

ecdf - Cumulative Distribution Function

```python
sns.displot(data=df, x='price', kind='ecdf', height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='mileage', kind='ecdf', height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```
