seaborn.scatterplot

seaborn.scatterplot(data=None, *, x=None, y=None, hue=None, size=None, style=None, palette=None, hue_order=None, hue_norm=None, sizes=None, size_order=None, size_norm=None, markers=True, style_order=None, legend='auto', ax=None, **kwargs)

Draw a scatter plot with possibility of several semantic groupings.

The relationship between x and y can be shown for different subsets of the data using the hue, size, and style parameters. These parameters control what visual semantics are used to identify the different subsets. It is possible to show up to three dimensions independently by using all three semantic types, but this style of plot can be hard to interpret and is often ineffective. Using redundant semantics (i.e. both hue and style for the same variable) can be helpful for making graphics more accessible.

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/refs/heads/main/merc.csv')
df.head()
```

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, y='price', x='year')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='year', y='mileage')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission', style='fuelType', s=100)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='mileage', y='price', hue='fuelType', s=100)
plt.show()
```

```python
sns.swarmplot(data=df, x=df['fuelType'], y=df['price'])
plt.show()
```
