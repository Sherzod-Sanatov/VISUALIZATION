Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Seaborn simplifies the process of creating complex visualizations by offering built-in themes, color palettes, and simplified methods for creating common plots.

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/refs/heads/main/museum_visitors.csv', index_col=0)
df.head()
```

```python
df.shape
```

```python
plt.plot(df)
plt.show()
```

```python
sns.lineplot(data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title("Muzeylarga tashriflar soni")
plt.xlabel('Sana')
plt.ylabel('Odamlar soni')
sns.lineplot(data=df)
plt.show()
```

it is possible to use x and y axes to reflect diagrams

```python
plt.figure(figsize=(10,6))
plt.title('Turli mezuylarga tashriflar')
plt.xlabel('Sana')
plt.ylabel('Odamlar soni')
plt.xticks(rotation=90)
sns.lineplot(x=df.index, y=df['Firehouse Museum'])
plt.show()
```

to rotate x core, use rotation (we are rotating to 90 degrees)

```python
plt.figure(figsize=(16,6))
plt.title("Muzeylarga tashriflar soni")
plt.xlabel('Sana')
plt.ylabel('Odamlar soni')
sns.lineplot(data=df)
plt.grid()
plt.xticks(rotation=90)
plt.show()
```

selecting one of lines in the plot

```python
plt.figure(figsize=(16,6))
plt.title("Muzeylarga tashriflar soni")
plt.xlabel('Sana')
plt.ylabel('Odamlar soni')
sns.lineplot(data=df['Chinese American Museum'])
plt.grid()
plt.xticks(rotation=90)
plt.show()
```

selecting two or more lines of the plot

```python
plt.figure(figsize=(16,6))
plt.title("Muzeylarga tashriflar soni")
plt.xlabel('Sana')
plt.ylabel('Odamlar soni')
sns.lineplot(data=df[['Firehouse Museum', 'America Tropical Interpretive Center']])
plt.grid()
plt.xticks(rotation=90)
plt.show()
```
