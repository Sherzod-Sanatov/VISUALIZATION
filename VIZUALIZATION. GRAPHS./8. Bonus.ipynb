```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/flight_delays.csv")
df.head()
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=df['Month'], y=df['AA'])
plt.title("America Airlines avia kompaniyasining o'rtacha kechikish vaqtlari")
plt.ylabel('Minutes')
plt.show()
```

```python
long_df = df.melt(id_vars=['Month'])
long_df.head()
```

```python
l_df = long_df.head(24)
l_df
```

```python
sns.catplot(data=l_df, x='Month', y='value', kind='bar', hue='variable')
plt.show()
```

```python
data = pd.melt(df, id_vars='Month', value_vars=['B6', 'EV','US'])
sns.catplot(data=data, x='Month', y='value', kind='bar', hue='variable')
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(data=df.set_index('Month'), cmap='Accent', annot=True)
plt.xlabel('Airline')
plt.show()
```
